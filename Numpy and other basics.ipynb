**Decorators let you alter the code without changing the original function, you just need to pass exceptions and an argument to otherwise execute the function with its parameters**

you must pass the same arguments into the inside function 

```python
def fix_name(function):
     
    def inside(name):                                           #must have same inputs as function being decorated
        
        if name=='ethan':                                       #exceptions for the decorator,
            return 'sorry your name is ethan'
        
        return function(name)                                  #otherwise execute the function with its parameter x
   
    return inside                                         
```

```python
@fix_name
def print_name(name):
    return 'Your name is ' + name
```

```python
print(print_name('john'))
print_name('ethan')
```

```python
def check_func(function):                                    #the parameter is function, the thing youre decorating
    
    def inside(x,y):
        
        if x ==0:
            return 'cant multiply by 0'                         #set perameters that you want
        if y==0:
            return 'cant multiply by 0'
        
       
        return function(x,y)                                    #if the exceptions arent met, execute the normal function with its parameters
   
    return inside                                               # return the inside function
```

```python
@check_func
def multiply(x,y):
    return (x*y)                                    #decorate the function with your "decorator function to alter its functionality"
```

```python
print(multiply(100,5))
print(multiply(0,100))
```

More Complex Decorator : *args, **kwargs

```python
import time
```

```python
def time_it(function):
    def decorator(*args, **kwargs):
        
        start= time.time()
        
        result= function(*args, **kwargs)                     #difference is that you set the function as a value 
        
        end= time.time()
        
        print('This function took ' + str( (end-start)*1000 + 100) + ' miliseconds')
        return result
    
    return decorator
```

```python
@time_it
def calc_square(numbers):
    result=[]
    for number in numbers:
        result.append(number*number)
    return result
```

```python
x= range(1,11)
calc_square(x)
```

When the interpreter detects the dependency of inner nested function on variable in the outer function, it stores the variable of the outer function even if you delete the outer function. Essentially, you can delete the outer function but the dependencies will remain available to the inner function

```python
def outer_func(msg):
    message= msg
    
    def inner_func():
        print(message)
    
    return inner_func


hi_func= outer_func('Hi')        #set a variable equal to a function 


del outer_func                    # i can delete the outerfunc 
hi_func()                         # but i initialized hi_func already so it stores the info and lets me continue

                                #functions are also objects 
```
